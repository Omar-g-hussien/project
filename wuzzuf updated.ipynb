```python
import pandas as pd 
import csv
import requests
from bs4 import BeautifulSoup as bs
```

```python
job_title = []
Company_name = []
job_location= []
job_description = []
```

```python
results = requests.get('https://wuzzuf.net/search/jobs/?q=&a=hpb')
```

```python
src = results.content
```

```python
print(src)
```

```python
soup = bs(src , 'lxml')
```

```python
soup
```

```python
jtitle = soup.find_all('h2' , {'class' : 'css-m604qf'})
jtitle[3].text
```

```python
Cname = soup.find_all('a' ,{'class' : 'css-17s97q8'})
```

```python
jlocation = soup.find_all('span' ,{'class' : 'css-5wys0k'})
```

```python
jdescription = soup.find_all('div' , {'class' : 'css-y4udm8'})
```

```python
for i in range(len(jtitle)):
    job_title.append(jtitle[i].text)
    Company_name.append(Cname[i].text)
    job_location.append(jlocation[i].text)
    job_description.append(jdescription[i].text)
    
print(job_title)
print(job_description)
print(job_location)
print(Company_name)
```

```python
from itertools import zip_longest
```

```python
file_list = [job_title, Company_name ,job_location , job_description]
exported = zip_longest(*file_list)
with open("D:\Data\wuzzuuf.csv" ,'w')as file:
    wr=csv.writer(file)
    wr.writerow(['jobtitle' , 'company_name' , 'location' , 'skills'])
    wr.writerows(exported)
```
