```python
# importing libraries 
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

```python
# inputting url
url = 'https://wuzzuf.net/search/jobs/?q=&a=hpb'
```

### create a client-based request to get the url

```python
client = urlopen(url)
```

```python
html = client.read()
```

```python
html
```

```python
client.close()
```

```python
soup= bs(html , 'html.parser')
```

```python
soup
```

```python
container = soup.find_all('div' , {'class' : 'css-1gatmva e1v1l3u10'})
len(container)
```

## bringing it together in csv file

```python
f = open('Downloads/wuzzuf1.csv', 'w')
header = 'job_title, Company_name, description\n'
f.write(header)
```

```python
for i in container:
    title = i.findAll('h2' , {'class' : 'css-m604qf'})
    job_title = title[0].text.strip()
    
    cname = i.findAll('a' , {'class' : 'css-17s97q8'})
    Company_name= cname[0].text.strip()
    
    jlocation = i.findAll('span' , {'class' : 'css-5wys0k'})
    job_location= jlocation[0].text.strip()
    
    type = i.findAll('span' , {'class' : 'css-1ve4b75 eoyjyou0'})
    description= type[0].text.strip()
    
    print(job_title)
    print(Company_name)
    print(job_location)
    print(description)
    print('------------------------------')
    
    f.write(job_title + ', ' + Company_name +', ' +description + '\n')
f.close()
```
